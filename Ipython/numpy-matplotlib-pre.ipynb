```python
%matplotlib inline 
#加入这一行才能在网页中显示图像 。。 不然调用外部容易卡死。。

import matplotlib.pyplot as plt
import numpy as np
from math import *


xt = np.arange(0,1,0.01)

y1 = np.sin(2*pi*xt)

plt.plot(xt, y1)

y2 = np.cos(2*pi*xt)


'''
    把图分成1x2的格子 并使用第1个格子
'''
plt.plot(xt, y2, color="red", linestyle='--')

plt.subplot(1,2,1)
plt.plot(xt, y1, label='First Line')
plt.subplot(1,2,2)
plt.plot(xt, y2, label='Second Line') 

'''必须有这两行 才能有legend的使用'''

plt.xlabel('X')
plt.ylabel('Y')

plt.title('Interesting Graph\nCheck it out')
plt.legend()


plt.show()
```
